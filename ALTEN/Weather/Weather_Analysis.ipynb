```python
import pandas as pd
```

```python
data = pd.read_table("Data/debilt.txt",skiprows=50,sep=',', na_values='     ')
```

```python
data.head()
```

```python
data.describe()
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
data.insert(loc=1,column='Year',value = data.YYYYMMDD.astype(str).str[:4].astype(int))
data.insert(loc=2,column='Month',value = data.YYYYMMDD.astype(str).str[4:6].astype(int))
data.insert(loc=3,column='Day',value = data.YYYYMMDD.astype(str).str[6:8].astype(int))
```

```python
year = [1993]
month = [8]#range(1,12)

mask1 = data.Year.isin(year)
mask2 = data.Month.isin(month)
total_mask = mask1&mask2

plt.figure(figsize=(12,6))
plt.suptitle(('Weather analysis in De Bilt in {} of {}').format(month,year))
plt.subplot(1, 2, 1)
plt.scatter(np.arange(total_mask.sum()),data["   TX"].loc[total_mask]/10,c='red')
plt.scatter(np.arange(total_mask.sum()),data["   TN"].loc[total_mask]/10,c='blue')
plt.title('Temperature')
plt.xticks(ticks = np.arange(total_mask.sum()),labels=data.YYYYMMDD[total_mask],rotation=90)
plt.ylabel('Temperature [C]')
plt.legend(['Max','Min'])

plt.subplot(1, 2, 2)
plt.scatter(np.arange(total_mask.sum()),data["   SQ"][total_mask]/10,c='orange')
plt.title('Amount of sun')
plt.xticks(ticks = np.arange(total_mask.sum()),labels=data.YYYYMMDD[total_mask],rotation=90)
plt.ylabel('Sun duration [h]')
plt.show()
```
